### Estimated Market Size Analysis of Electric Vehicles in the United States

```python
## Calculate the number of EVs registered each year
ev_registrations_count = df['Model Year'].value_counts().sort_index()
```

```python
ev_registrations_count
```

```python
## Importing required libraries 
from scipy.optimize import curve_fit
```

```python
## Filtering the dataset to include years with complete data, assuming 2023 is the last complete year
filtered_years = ev_registrations_count[ev_registrations_count.index<=2023]
```

```python
filtered_years
```

```python
## Defining an Exponential Growth Function
def exp_growth(x, a, b):
    return a * np.exp(b * x)
```

```python
## Preparing Data for Curve Fitting
x_data = filtered_years.index - filtered_years.index.min()
```

```python
y_data = filtered_years.values
```

```python
x_data
```

```python
y_data
```

```python
## Applying Curve Fitting
params, covariance = curve_fit(exp_growth,x_data,y_data)
```

```python
## params: The best-fit values [a, b].
params
```

```python
## covariance: A matrix that estimates the uncertainty in these parameters.
covariance
```

#### Forecasting Future EV Counts

```python
forecast_years = np.arange(2024,2024+6) - filtered_years.index.min()
```

```python
forecast_years
```

```python
forecasted_values = exp_growth(forecast_years, *params)
```

```python
forecasted_values
```

```python
forecasted_evs = dict(zip(forecast_years + filtered_years.index.min(), forecasted_values))
```

```python
print(forecasted_evs)
```
